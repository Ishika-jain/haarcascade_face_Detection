```python
frame = cv2.imread("11.jpg")
```

```python
plt.imshow(frame)
```

```python
rgbframe = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgbframe)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python

faces = classifier.detectMultiScale(frame)

```

```python
face = faces[0]
```

```python
x,y,h,w = face
```

```python
face = rgbframe[y:y+h, x:x+w]
```

```python
plt.imshow(face)
```

```python
out = cv2.rectangle(rgbframe, (x,y), (x+w,y+h), (0,0,255), 10)
```

```python
plt.imshow(out)
```
